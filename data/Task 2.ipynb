# Function to merge all the files

```python
# Acessing the filepath
file_path = "D:\Forage\Quantium\sw\quantium-starter-repo\data"
file_list = os.listdir(file_path)
file_list
```

```python
# Merging files together
data = pd.concat(pd.read_csv(f"D:\Forage\Quantium\sw\quantium-starter-repo\data\{f}") for f in file_list)
data
```

```python
data.isnull().sum()
```

```python
uniques_products_counts =data["product"].value_counts().reset_index()
uniques_products_counts
```

```python
uniques_products = uniques_products_counts["product"].to_list()
uniques_products
```

```python
unwanted_products = data.loc[data["product"]!="pink morsel"]
unwanted_products
```

```python
data = pd.concat([data,unwanted_products]).drop_duplicates(keep= False)
data
```

```python
data.isnull().sum()
```

```python
data.dtypes 
```

```python
# as price and the quantity are different d types i am converting price to int
data.rename(columns={"price":"price in $"}, inplace= True)
data.head(5)
```

```python
data["price in $"] = data["price in $"].str.replace("$","")
data
```

```python
data["price in $"] = data["price in $"].astype(np.float64)
```

```python
data["Total Sales"] = data["price in $"] * data["quantity"]
data
```

```python
data.rename(columns={"Total Sales":"Total Sales in $"}, inplace= True)
data.head(5)
```

```python
#rearranging the row numbers

data = data.reset_index(drop=True)
data
```
